# You have to design a Food Ordering app for a restaurant

```python
import random

# Define a class for food items
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = random.randint(1000, 9999)
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

    def __str__(self):
        return f"FoodID: {self.food_id}, Name: {self.name}, Quantity: {self.quantity}, Price: INR {self.price}, Discount: {self.discount}%, Stock: {self.stock}"

# Define a class for users
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, food_items):
        self.orders.append(food_items)

    def view_order_history(self):
        if not self.orders:
            print("You have no order history.")
        else:
            print("Order History:")
            for order in self.orders:
                print(order)

    def update_profile(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password

# Define a class for the restaurant with food items
class Restaurant:
    def __init__(self):
        self.food_menu = []

    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        self.food_menu.append(food_item)

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for item in self.food_menu:
            if item.food_id == food_id:
                item.name = name
                item.quantity = quantity
                item.price = price
                item.discount = discount
                item.stock = stock
                print(f"Food item with FoodID {food_id} has been updated.")
                return
        print(f"Food item with FoodID {food_id} not found.")

    def view_food_menu(self):
        if not self.food_menu:
            print("Food menu is empty.")
        else:
            print("Food Menu:")
            for item in self.food_menu:
                print(item)

    def remove_food_item(self, food_id):
        for item in self.food_menu:
            if item.food_id == food_id:
                self.food_menu.remove(item)
                print(f"Food item with FoodID {food_id} has been removed.")
                return
        print(f"Food item with FoodID {food_id} not found.")

# Sample usage of the classes
restaurant = Restaurant()

# Adding some food items to the menu
restaurant.add_food_item("Tandoori Chicken (4 pieces)", "4 pieces", 240, 0, 100)
restaurant.add_food_item("Vegan Burger (1 Piece)", "1 piece", 320, 10, 50)
restaurant.add_food_item("Truffle Cake (500gm)", "500gm", 900, 5, 20)


# Admin functionalities
admin_password = "1234567890"  # Password for admin access (not secure in this example)
if input("Enter admin password: ") == admin_password:
    while True:
        print("\nAdmin Menu:")
        print("1. Add new food item")
        print("2. Edit food item")
        print("3. View food menu")
        print("4. Remove food item")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter food name: ")
            quantity = input("Enter quantity: ")
            price = float(input("Enter price: "))
            discount = float(input("Enter discount percentage: "))
            stock = int(input("Enter stock: "))
            restaurant.add_food_item(name, quantity, price, discount, stock)
            print("Food item has been added.")

        elif choice == "2":
            food_id = int(input("Enter FoodID to edit: "))
            name = input("Enter food name: ")
            quantity = input("Enter quantity: ")
            price = float(input("Enter price: "))
            discount = float(input("Enter discount percentage: "))
            stock = int(input("Enter stock: "))
            restaurant.edit_food_item(food_id, name, quantity, price, discount, stock)

        elif choice == "3":
            restaurant.view_food_menu()

        elif choice == "4":
            food_id = int(input("Enter FoodID to remove: "))
            restaurant.remove_food_item(food_id)

        elif choice == "5":
            break

        else:
            print("Invalid choice. Please try again.")

# End of admin functionalities

```

```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, food_menu):
        print("\nFood Menu:")
        for i, item in enumerate(food_menu, start=1):
            print(f"{i}. {item['name']} [{item['quantity']}] [INR {item['price']}]")

        selected_items = []
        while True:
            try:
                choice = int(input("\nEnter the number(s) of the item(s) you want to order (0 to finish): "))
                if choice == 0:
                    break
                elif choice < 1 or choice > len(food_menu):
                    print("Invalid choice. Please select a valid item.")
                else:
                    selected_items.append(food_menu[choice - 1])
            except ValueError:
                print("Invalid input. Please enter a number.")

        if selected_items:
            print("\nSelected Items:")
            total_cost = 0
            for item in selected_items:
                print(f"{item['name']} [{item['quantity']}] [INR {item['price']}]")
                total_cost += item['price']
            print(f"\nTotal Cost: INR {total_cost}")
            place_order = input("Do you want to place this order? (yes/no): ").strip().lower()
            if place_order == "yes":
                self.orders.append(selected_items)
                print("Order placed successfully!")
            else:
                print("Order not placed.")
        else:
            print("No items selected for the order.")

    def view_order_history(self):
        if not self.orders:
            print("You have no order history.")
        else:
            print("\nOrder History:")
            for i, order in enumerate(self.orders, start=1):
                print(f"Order {i}:")
                total_cost = sum(item['price'] for item in order)
                for item in order:
                    print(f"{item['name']} [{item['quantity']}] [INR {item['price']}]")
                print(f"Total Cost: INR {total_cost}")
                print()

    def update_profile(self):
        print("\nUpdate Profile:")
        self.full_name = input("Full Name: ")
        self.phone_number = input("Phone Number: ")
        self.email = input("Email: ")
        self.address = input("Address: ")
        self.password = input("Password: ")

def main():
    food_menu = [
        {"name": "Tandoori Chicken", "quantity": "4 pieces", "price": 240},
        {"name": "Vegan Burger", "quantity": "1 Piece", "price": 320},
        {"name": "Truffle Cake", "quantity": "500gm", "price": 900},
    ]

    user = None

    while True:
        print("\nWelcome to the Food Ordering App!")
        print("1. Register")
        print("2. Log In")
        print("3. Exit")
        choice = input("Select an option: ")

        if choice == "1":
            if user is None:
                full_name = input("Full Name: ")
                phone_number = input("Phone Number: ")
                email = input("Email: ")
                address = input("Address: ")
                password = input("Password: ")
                user = User(full_name, phone_number, email, address, password)
                print("Registration successful!")
            else:
                print("You are already logged in. Please log out to register a new account.")
        elif choice == "2":
            if user is None:
                print("You are not registered. Please register first.")
            else:
                entered_password = input("Enter Password: ")
                if entered_password == user.password:
                    while True:
                        print("\nUser Menu:")
                        print("1. Place New Order")
                        print("2. Order History")
                        print("3. Update Profile")
                        print("4. Log Out")
                        user_choice = input("Select an option: ")
                        if user_choice == "1":
                            user.place_order(food_menu)
                        elif user_choice == "2":
                            user.view_order_history()
                        elif user_choice == "3":
                            user.update_profile()
                        elif user_choice == "4":
                            print("Logged out.")
                            break
                        else:
                            print("Invalid choice. Please try again.")
                else:
                    print("Incorrect password. Please try again.")
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
